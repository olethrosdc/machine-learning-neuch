```python
T = 1000
theta = 0.9
import numpy as np
```

```python
X = np.random.choice(2, p=[1-theta, theta], size=T)
np.mean(X)
```

```python
N = 10000
v = np.zeros(N)
for n in range(N):
    idx = np.random.choice(T, size=T, replace = True)
    v[n] = np.mean(X[idx])
```

```python
import matplotlib.pyplot as plt
plt.hist(v)
```

```python
v = np.sort(v)
```
